```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linia(x):
    return x*x+5
```

```python
x=np.linspace(-1, 1, 20)
x1=np.linspace(-6, 6, 40)
x2=np.linspace(0, 5, 30)

plt.plot(x,linia(x),label="Od -1 do 1")
plt.plot(x1,linia(x1),label="Od -6 do 6")
plt.plot(x2,linia(x2),label="Od 0 do 5")

plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.title("Na≈Ço≈ºony wykres tej samej funkcji dla r√≥≈ºnych warto≈õci")
plt.legend()
```
